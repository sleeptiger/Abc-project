## Figure와 그래프 요소
> - matplotlib 에서 **그래프를 구성하는 가장 큰 요소는 `figure`입니다. = 액자**
> - `axes`는 보통 plot으로 생각하는 **하나의 그래프**입니다. (액자속 사진)
> - 하나의 `figure`에는 여러개의 `axes`를 가질 수 있고, `axes`는 각각 x,y축(axis)가집니다.
> - `figure`를 별도로 지정준주지 않으면 matplotib이 자동으로 생성된다.

```python
plt.plot(x, y, color="darkblue", linewidth=3, linestyle="--", marker="o", ms=15, mec="orange", mew=3, mfc="yellow")
plt.show()
```

```python
#figure 크기 지정 (figsize=(가로, 세로))
plt.figure(figsize=(15,5))
plt.plot(x,y, x, y, color="darkblue", linewidth=3, linestyle="--", marker="o", ms=15, mec="orange", mew=3, mfc="yellow")
plt.show()
```

**plot 범위 설정**
> - `xlim`과 `ylim`을 사용해서 **x축과 y축에서 표현될 최소값, 최대값을 설정할 수 있습니다.
> - 기본 사용법 : `plt.xlim(최소, 최대), plt.ylim(최소, 최대)`


축(axis), 제목(title) 이름
> - `xlabel`, `ylabel`, `title`을 사용해서 그래프의 가독성을 높일 수 있다.

```python
plt.figure(figsize=(15, 5))
plt.rc('font', size=20)     # 폰트 사이즈 변경
plt.plot(x, y, color="darkblue", linewidth=3, linestyle="--", marker="o", ms=15, mec="orange", mew=3, mfc="yellow")
plt.xlabel('X 축')
plt.ylabel('Y 축')
plt.title('아주 멋진 그래프')
plt.show()
```

**tick 설정**
> - 그래프에서 축에 표시되는 지점을 `tick`이라고 한다.
> - 이 `tick`에 나와있는 숫자 혹은 글자를 `tick label`이라고 합니다.
> - **보여지고 싶은 간격**을 설정할때 사용합니다.

```python
plt.plot(x, y, color="darkblue", linewidth=3, linestyle="--", marker="o", ms=15, mec="orange", mew=3, mfc="yellow")
plt.xticks(np.linspace(0, 80, 5)) # 0~80까지 5등분 한다.
plt.yticks(np.linspace(0, 15, 3)) # 0~15까지 3등분 한다.
plt.show()
```

멀티 라인 그래프
> - plt.plot을 연속적으로 사용하여 하나의 figure 여러개 그래프가 함계 그려집니다.

```python
x = np.linspace(-np.pi, np.pi, 200)
y1, y2 = np.cos(x), np.sin(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

### **바(bar)**


- X나 Y중 하나의 값이 카테고리함 일 경우 바(`bar`)형태 차트가 효과적인 시각화 방법
- 가로방향으로 그리려면 `bar`로 그릴 수 있습니다.
- 기본 사용법
    - `plt.bar(x,y, color=['색상'], width='폭')`

```python
# 데이터 생성
x = np.arange(5)
y = [80, 70, 40, 90, 60]
xlabel = ['국어', '수학', '영어', '과학', '미술', '체육']
```

```python
plt.bar(x,y)
plt.xticks(x, xlabel) # X (value)를 X라벨로 바꾼다
plt.ylabel('점수')
plt.show()
```

```python
# 가로 막대 그래프
plt.barh(x,y)
plt.xticks(x, xlabel)
plt.ylabel('점수')
plt.show()
```

### [bar] 여러개 그리기 (나란한 형태로)
    - 바의 폭 크기를 지정한 후, 수동으로 다음 그려질 그래프 위치를 지정해 주어야 합니다. (어려움)

```python
x = np.arange(5)
women = [80, 70, 40, 90, 60]
man = [70, 80, 45, 75, 80]
xlabel=['국어', '수학', '영어', '과학', '미술', '체육']
```

```python
bar_width = 0.35
plt.bar(x, women, width=bar_width, label='여자')
plt.bar(x + bar_width, man, width=bar_width, label='남자') # (기준값 + 폭넓이, value)
plt.xticks(x+ bar_width/2, xlabel)
plt.ylabel('점수')
plt.legend(loc='upper right')
plt.show()
```

### [bar]여러개 그리기(Stacked 형태_
>-  `bottom` 옵션을 사용해 겹치진 바 차트도 만들 수 있다. 

```python
bar_width = 0.35
plt.bar(x, women, width=bar_width, label='여자', bottom=man)
plt.bar(x, man, width=bar_width, label='남자')
plt.xticks(x, xlabel)
plt.ylabel('점수')
plt.legend('upper right')
plt.show()
```

### 파이(pie)
> - 카테고리 별 값의 상대적인 비교를 할 경우, `pie` 차트가 선호
    - `plt.pie(카테고리 값 color=['색상'], width='폭', explode='돌출 옵션')`

```python
# 데이터 생성
labels = ['애플', '삼성', '오포', '샤오미', '기타']
sizes = [16, 21, 16, 14, 33]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'lightgray']
```

```python
plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, colors=colors)
plt.title('휴대폰 시장 점유율')
plt.show()
```

```python
explode = (0, 0.1, 0, 0, 0)

plt.figure(figsize=(5,5))
plt.pie(sizes, explode=explode,labels=labels, colors=colors, autopct='%.1f%%', shadow=True)
plt.title('휴대폰 시장 점유율')
plt.show()
```

### 히스토그램(histogram
- 빈도 분포가 어떻게 되어 있는지 확인을 위해서는, `hist`차트가 선호된다.
- 기본 사용법
    - `plt.hist(발생 빈도가 있는 값)`

```python
x = np.random.randn(2000)
plt.hist(x)
plt.show()
```

```python
# bin 크기 지정
plt.hist(x, bins=100)
plt.show()
```

```python
# 누적 분포형
plt.hist(x, bins=100, cumulative=True)
plt.show()
```

**이미지(imshow)**

3D 그래프(surface)
- 3차 그래프는 일반 figure와 달리, 3차원 전용 `axes`를 생성해서 입체적으로 표현한다. 
- 기본 사용법
    - `axes.plot_surface(x,y,z)`

```python
# 3차원 데이터 생성
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(50,150,24)
y = np.linspace(50,150,24)

x, y = np.meshgrid(x, y)     # x 와 y 의 shape 을 합칩니다.
z = (x-100)**2/y             # z 값 생성 (2차원 배열)
```

```python
fig = plt.figure(figsize=(10,5))
axes = Axes3D(fig)
axes.plot_surface(x, y, z, cmap='coolwarm')
plt.title('3D 그래프 시각화')
plt.show()
```

> ### **matplotlib 그래프를 seabor하에서 표현하기**
> - `matplotlib`으로 그리는 그래프는 `seaborn`에서 똑같이 연출 가능
> - `matplotlib`과 `seaborn`을 섞어서도 사용 가능

```python
#데이터 로드
penguins = sns.load_dataset('penguins')
penguins.head()
```

> **matplotlib 으로 그리기**

```python
fig, ax = plt.subplots(ncols=2, figsize=(10,5))

#axes[0] = matplotlib
islands = penguins['island'].unique()
color= 0
for island in islands:
    ax[0].scatter(penguins['bill_length_mm'].loc[penguins['island']==island],
                    penguins['bill_depth_mm'].loc[penguins['island']==island],
                    c=f'C{color}', label=island)
    color = color + 1
ax[0].set_xlabel('bill_length_mm')
ax[0].set_ylabel('bill_depth_mm')

plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10,5))

#axes[0] = matplotlib
islands = penguins['island'].unique()
color= 0
for island in islands:
    ax[0].scatter(penguins['bill_length_mm'].loc[penguins['island']==island],
                    penguins['bill_depth_mm'].loc[penguins['island']==island],
                    c=f'C{color}', label=island)
    color = color + 1
ax[0].set_xlabel('bill_length_mm')
ax[0].set_ylabel('bill_depth_mm')

sns.scatterplot('bill_length_mm', 'bill_depth_mm', hue='island', data=penguins, ax=ax[1])

plt.show()
```

> **matplotlib + seaborn혼합해서 사용하기**
> - matplotlib과 seaborn은 자유로게 섞어서 사용 가능
> - 외쪽 그래프 : matplotlib 산점도 + seaborn 추세선
> - 오른쪽 그래프 : seaborn 산점도 + matplotlib 중심점

```python
fig, axes = plt.subplots(ncols=2, figsize=(10,5))

#axes[0] = matplotlib
islands = penguins['island'].unique()
color= 0
for island in islands:
    axes[0].scatter(penguins['bill_length_mm'].loc[penguins['island']==island],
                    penguins['bill_depth_mm'].loc[penguins['island']==island],
                    c=f'C{color}', label=island)
    color = color + 1

    #seaborn 추세선
    sns.regplot('bill_length_mm', 'bill_depth_mm', data=penguins.loc[penguins['island']==island],
                scatter=False, ax=axes[0])
    
axes[0].set_xlabel('bill_length_mm')
axes[0].set_ylabel('bill_depth_mm')

sns.scatterplot('bill_length_mm', 'bill_depth_mm', hue='island',data=penguins, 
                ax=axes[1])

for island in islands:
    #matplotlib중심점
    axes[1].scatter(penguins['bill_length_mm'].loc[penguins['island'] == island].mean(),
                    penguins['bill_depth_mm'].loc[penguins['island']== island].mean(),
                    c='black', marker='*', s=500)
plt.show()
```
